```python
from tkinter import *

root = Tk()
root.title("Calculator")
# take input
n = Entry(root, width=50, borderwidth = 5)
n.grid(row =0, column = 0, columnspan= 5, padx = 10, pady = 10)

def button_click(number):
    #n.delete (0, END)  #delete the entry from the screen
    # it is deleting the first entry when puting the second.
    
    #n.insert(0, number)  #put the result on the screen/ entry bar  
    # this here is casuing problem while input. 567 is being written as 765
    current = n.get()
    n.delete(0, END)
    n.insert(0, str(current) + str(number))
    
def button_clear():
    n.delete(0, END)
    
def button_add():
    firstnum = n.get()
    global f_num
    global math
    math = "Addition"
    f_num = int(firstnum)
    n.delete(0, END)
    
def button_equal():
    secondnum = n.get()
    n.delete(0, END)
    if math == "Addition":
        n.insert(0, f_num + int(secondnum))
    if math == "Subtract":
        n.insert(0, f_num - int(secondnum))
    if math == "Multiply":
        n.insert(0, f_num * int(secondnum))
    if math == "Division":
        n.insert(0, f_num / int(secondnum))
    
    
def button_sub():
    firstnum = n.get()
    global f_num
    global math
    math = "Subtract"
    f_num = int(firstnum)
    n.delete(0, END)
    
def button_mul():
    firstnum = n.get()
    global f_num
    global math
    math = "Multiply"
    f_num = int(firstnum)
    n.delete(0, END)
    
def button_div():
    firstnum = n.get()
    global f_num
    global math
    math = "Division"
    f_num = int(firstnum)
    n.delete(0, END)
    
```

```python
#create buttons
#button0 = Button(root, text = "0", padx = 40, pady = 20, command = button_click)
#here we can not pass parameters to the function, to send the parameters we have to use lambda funtion
button0 = Button(root, text = "0", padx = 40, pady = 20, command = lambda: button_click(0))
button1 = Button(root, text = "1", padx = 40, pady = 20, command = lambda: button_click(1))
button2 = Button(root, text = "2", padx = 42, pady = 20, command = lambda: button_click(2))
button3 = Button(root, text = "3", padx = 40, pady = 20, command = lambda: button_click(3))
button4 = Button(root, text = "4", padx = 40, pady = 20, command = lambda: button_click(4))
button5 = Button(root, text = "5", padx = 42, pady = 20, command = lambda: button_click(5))
button6 = Button(root, text = "6", padx = 40, pady = 20, command = lambda: button_click(6))
button7 = Button(root, text = "7", padx = 40, pady = 20, command = lambda: button_click(7))
button8 = Button(root, text = "8", padx = 42, pady = 20, command = lambda: button_click(8))
button9 = Button(root, text = "9", padx = 40, pady = 20, command = lambda: button_click(9))
buttonadd = Button(root, text = "+", padx = 39, pady = 20, command = button_add)
buttonsub = Button(root, text = "-", padx = 41, pady = 20, command = button_sub)
buttonmul = Button(root, text = "*", padx = 41, pady = 20, command = button_mul)
buttondiv = Button(root, text = "/", padx = 39, pady = 20, command = button_div)
buttonequals = Button(root, text = "=", padx = 40, pady = 20, command = button_equal)
buttonclear = Button(root, text = "clear", padx = 35, pady = 20, command = button_clear)

# put the buttons on screen
button7.grid(row = 1, column = 0)
button8.grid(row = 1, column = 1)
button9.grid(row = 1, column = 2)
button4.grid(row = 2, column = 0)
button5.grid(row = 2, column = 1)
button6.grid(row = 2, column = 2)
button1.grid(row = 3, column = 0)
button2.grid(row = 3, column = 1)
button3.grid(row = 3, column = 2)
button0.grid(row = 4, column = 0)
buttonadd.grid(row = 4, column = 4)
buttonsub.grid(row = 3, column = 4)
buttonmul.grid(row = 2, column = 4)
buttondiv.grid(row = 1, column = 4)
buttonequals.grid(row = 4, column = 2)
buttonclear.grid(row =4 , column =1)

root.mainloop()
```
